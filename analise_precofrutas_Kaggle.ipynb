```python
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Apples', 'Maçã', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Apricots', 'Damascos', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Bananas', 'Bananas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Blackberries', 'Amoras', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Blueberries', 'Mirtilos', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Cantaloupe', 'Melões Cantalupo', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Cherries', 'Cerejas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Clementines', 'Tangerinas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Grapefruit', 'Toronjas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Grapes', 'Uvas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Honeydew', 'Melões Honeydew', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Kiwi', 'Kiwi', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Mangoes', 'Mangas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Nectarines', 'Nectarinas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Oranges', 'Laranjas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Papaya', 'Mamão', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Peaches', 'Pêssegos', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Pears', 'Peras', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Pineapple', 'Abacaxi', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Plum', 'Ameixas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Pomegranate', 'Romãs', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Raspberries', 'Framboesas', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Strawberries', 'Morangos', df_frutas['Fruta'])
df_frutas['Fruta'] = np.where(df_frutas['Fruta'] == 'Watermelon', 'Melancia', df_frutas['Fruta'])


```

<h3>As Duas Frutas De Maior Preço</h3>

<h3> Fruta Com Menor Preço </h3>

```python
df_frutas[df_frutas["Preco_EquivalenteAoPeso"] == df_frutas["Preco_EquivalenteAoPeso"].min()]
```

<h2> ANALISE DOS PREÇOS DAS FRUTAS </h2>

```python
df_frutas = df_frutas.sort_values(by=["Preco_EquivalenteAoPeso"])
```

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=df_frutas, kind="bar",
    x="Fruta", y="Preco_EquivalenteAoPeso",
    palette="dark", ci=None
);
#g.despine(left=True)
g.set_axis_labels("Frutas", "Preço por Unidade de Medida (Lb = Libras)")
g.fig.suptitle('Quais São As Frutas Com Maior, ou Menor, Preço em Relação Ao Seu Peso', size=15)
#g.set_xlabel('Frutas', fontdict={'size': 15})
g.set_xticklabels(

    rotation=90,
     size = 8
)
plt.show();
```

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=df_frutas, kind="bar",
    x="Fruta", y="Rendimento",
    palette="dark", ci=None
);
g.set_axis_labels("Frutas", "Rendimento da Fruta")
g.fig.suptitle('Quais São As Frutas Com Maior, ou Menor, Rendimento', size=15)
g.set_xticklabels(

    rotation=90,
     size = 8
)
plt.show();
```

```python
df_frutas['Preco_EmRelacaoAoRendimentoDaFruta'] = df_frutas["Preco_EquivalenteAoPeso"]/df_frutas["Rendimento"]
df_frutas.head()
```

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=df_frutas, kind="bar",
    x="Fruta", y="Preco_EmRelacaoAoRendimentoDaFruta",
    palette="dark", ci=None
);

g.set_axis_labels("Frutas", "Preço por Unidade de Libras / Rendimento")
g.fig.suptitle('Quais São As Frutas Com Maior, ou Menor, Preço em Relação Ao Seu Rendimento', size=15)
g.set_xticklabels(
    #g.get_xticklabels(),
    rotation=90,
     size = 8
)
plt.show();
```

<h3> Frutas com Menor e Maior Preço, em relação ao seu rendimento, respectivamente </h3>


```python
df_frutas[(df_frutas["Preco_EmRelacaoAoRendimentoDaFruta"] == df_frutas["Preco_EmRelacaoAoRendimentoDaFruta"].max()) | (df_frutas["Preco_EmRelacaoAoRendimentoDaFruta"]==df_frutas["Preco_EmRelacaoAoRendimentoDaFruta"].min())].head()
```

```python
sns.set_theme(style="whitegrid")
g = sns.catplot(
    data= df_frutas[(df_frutas.Fruta == "Bananas") | (df_frutas.Fruta == "Romãs")] , kind="bar",
    x="Fruta", y="Preco_EmRelacaoAoRendimentoDaFruta",
    palette="dark", ci=None
);

g.set_axis_labels("Frutas", "Preço por Unidade de Libras / Rendimento")
g.fig.suptitle('Frutas Com Menor e Maior Preço em Relação Ao Seu Rendimento', size=15)
g.set_xticklabels(
    #g.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
)
plt.show();
```

```python

sns.set_theme(style="whitegrid")
g = sns.catplot(
    data=df_frutas.sort_values(by=["Preco_EmRelacaoAoRendimentoDaFruta"]), kind="bar",
    x="Fruta", y="Preco_EmRelacaoAoRendimentoDaFruta",
    palette="dark", ci=None
);

g.set_axis_labels("Frutas", "Preço por Unidade de Libras / Rendimento")
g.fig.suptitle('Frutas Com Menor e Maior Preço em Relação Ao Seu Rendimento', size=15)
g.set_xticklabels(
    #g.get_xticklabels(),
    rotation=90,
     size = 8
)
plt.show();
```
